```python
from flask import Flask, request, send_file, jsonify, send_from_directory, abort
from flask_cors import CORS
from gtts import gTTS
from werkzeug.middleware.proxy_fix import ProxyFix
import io
import os
import subprocess
import time
import logging
import librosa
import gc
import torch
import cv2
import glob
import shutil
import base64
import concurrent.futures
import numpy as np
import mediapipe as mp
import requests
import math
from tqdm import tqdm
from werkzeug.utils import secure_filename
from moviepy.editor import VideoFileClip, ImageSequenceClip, AudioFileClip


# Check GPU availability and optimize settings
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
if device.type == 'cuda':
    # Print GPU info
    !nvidia-smi
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
    # Optimize GPU settings
    torch.cuda.set_device(0)
    torch.backends.cudnn.benchmark = True
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.enabled = True
    torch.cuda.empty_cache()
else:
    print("No GPU available, using CPU instead. Processing will be slower.")

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)



# Initialize Flask app
app = Flask(__name__)
# Add this after creating the Flask app
app.wsgi_app = ProxyFix(app.wsgi_app, x_proto=1, x_host=1)
CORS(app, resources={
    r"/*": {
        "origins": "*",
        "methods": ["GET", "POST", "OPTIONS"],
        "allow_headers": ["Content-Type", "Authorization"],
        "supports_credentials": True
    }
})

global_timestamp = None
# Ensure paths exist
UPLOAD_FOLDER = "/content"
INPUT_FACE_VIDEO = os.path.join(UPLOAD_FOLDER, "face_detect_inputs")
FACE_FRAMES_FOLDER = os.path.join(UPLOAD_FOLDER, "face_frames")
PROCESSED_VIDEOS_FOLDER = os.path.join(UPLOAD_FOLDER, "processed_videos")

# Define paths
BASE_PATH = "/content/Wav2Lip-GFPGAN"
WAV2LIP_PATH = os.path.join(BASE_PATH, "Wav2Lip-master")
OUTPUT_PATH = os.path.join(BASE_PATH, "outputs")
CODEFORMER_PATH = "/content/CodeFormer"

# Create output directory if it doesn't exist
os.makedirs(OUTPUT_PATH, exist_ok=True)

# Ensure folders exist
os.makedirs(INPUT_FACE_VIDEO, exist_ok=True)
os.makedirs(FACE_FRAMES_FOLDER, exist_ok=True)
os.makedirs(PROCESSED_VIDEOS_FOLDER, exist_ok=True)

# Add this code block here - CodeFormer requirements check
try:
    # Check CodeFormer requirements
    if not os.path.exists(CODEFORMER_PATH):
        logger.warning("CodeFormer directory not found. Enhancement features will be disabled.")
    elif not os.path.exists(os.path.join(CODEFORMER_PATH, 'weights/CodeFormer/codeformer.pth')):
        logger.warning("CodeFormer weights not found. Enhancement features will be disabled.")
    else:
        # Create CodeFormer directories if they don't exist
        os.makedirs(os.path.join(CODEFORMER_PATH, 'weights/CodeFormer'), exist_ok=True)
        os.makedirs(os.path.join(CODEFORMER_PATH, 'weights/retinaface_resnet50'), exist_ok=True)
        logger.info("CodeFormer requirements checked and directories created successfully")
except Exception as e:
    logger.error(f"Error checking CodeFormer requirements: {e}")

mp_face_mesh = mp.solutions.face_mesh
mp_face_detection = mp.solutions.face_detection
face_mesh = mp_face_mesh.FaceMesh(refine_landmarks=True)
face_detection = mp_face_detection.FaceDetection()

def process_enhanced_video(input_path, output_path):
    """Process video with CodeFormer enhancement"""
    try:
        # Check for CodeFormer requirements
        weights_path = os.path.join(CODEFORMER_PATH, 'weights/CodeFormer/codeformer.pth')
        if not os.path.exists(weights_path):
            raise Exception(f"CodeFormer weights not found at {weights_path}")

        if not os.path.exists(CODEFORMER_PATH):
            raise Exception(f"CodeFormer directory not found at {CODEFORMER_PATH}")

        # Store original directory
        original_dir = os.getcwd()
        os.chdir(CODEFORMER_PATH)

        start_time = time.time()

        # Settings
        w = 0.7  # CodeFormer fidelity
        upscale = 1  # No upscaling for speed
        target_fps = 40  # Fixed processing and output FPS

        print("\n===== FAST MODE: CodeFormer Video Enhancement =====")
        print(f"Settings: w={w}, fixed 40 FPS output (no interpolation)")

        # Create directories
        os.makedirs('inputs/whole_imgs', exist_ok=True)
        os.makedirs('results', exist_ok=True)

        # Extract audio
        print("Extracting audio...")
        video_clip = VideoFileClip(input_path)
        audio_clip = video_clip.audio
        temp_audio_path = "temp_audio.mp3"
        if audio_clip is not None:
            audio_clip.write_audiofile(temp_audio_path, verbose=False, logger=None)
        video_clip.close()

        # Get video properties
        cap = cv2.VideoCapture(input_path)
        original_fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        duration = total_frames / original_fps

        # Calculate frame sampling rate
        total_output_frames = int(duration * min(original_fps, target_fps))
        frame_interval = total_frames / total_output_frames if original_fps > target_fps else 1

        print(f"\nVideo properties:")
        print(f"Resolution: {width}x{height}")
        print(f"Original FPS: {original_fps}")
        print(f"Target FPS: {min(original_fps, target_fps)}")
        print(f"Total frames: {total_frames}")
        print(f"Duration: {duration:.2f}s")
        print(f"Sampling interval: {frame_interval:.2f} frames")

        # Extract frames
        print("\nExtracting frames...")
        frame_count = 0
        saved_count = 0
        next_frame_to_save = 0

        with tqdm(total=total_output_frames, desc="Extracting frames") as pbar:
            while True:
                ret, frame = cap.read()
                if not ret:
                    break

                if frame_count >= next_frame_to_save:
                    # Resize frame if needed
                    if width > 1280 or height > 1280:
                        scale_factor = 1280 / max(width, height)
                        new_width = int(width * scale_factor)
                        new_height = int(height * scale_factor)
                        frame = cv2.resize(frame, (new_width, new_height))

                    frame_path = f'inputs/whole_imgs/{saved_count:08d}.png'
                    cv2.imwrite(frame_path, frame)
                    saved_count += 1
                    next_frame_to_save += frame_interval
                    pbar.update(1)

                frame_count += 1

        cap.release()
        print(f"\nSuccessfully extracted {saved_count} frames")

        # Create inference script
        inference_script = os.path.join(CODEFORMER_PATH, 'fast_inference.py')
        with open(inference_script, 'w') as f:
            script_content = '''
import os
import argparse
import cv2
import glob
import torch
from torchvision.transforms.functional import normalize
from facexlib.detection import RetinaFace
from facexlib.utils.face_restoration_helper import FaceRestoreHelper
from basicsr.utils.registry import ARCH_REGISTRY
from basicsr.utils import img2tensor, tensor2img
import numpy as np
from tqdm import tqdm

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('-i', '--input', type=str)
    parser.add_argument('-o', '--output', type=str)
    parser.add_argument('-w', '--fidelity_weight', type=float, default=0.7)
    parser.add_argument('-s', '--upscale', type=int, default=1)
    args = parser.parse_args()

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Load CodeFormer
    net = ARCH_REGISTRY.get('CodeFormer')(dim_embd=512, codebook_size=1024, n_head=8, n_layers=9,
                                        connect_list=['32', '64', '128', '256']).to(device)

    checkpoint = torch.load('{0}')['params_ema']
    net.load_state_dict(checkpoint)
    net.eval()

    # Setup face helper
    face_helper = FaceRestoreHelper(
        upscale_factor=args.upscale,
        face_size=512,
        use_parse=False,
        device=device,
        model_rootpath='{1}'
    )

    # Process images
    paths = sorted(glob.glob(os.path.join(args.input, '*.png')))
    os.makedirs(os.path.join(args.output, 'final_results'), exist_ok=True)

    print(f"Found {{len(paths)}} images to process")

    with tqdm(total=len(paths), desc="Processing frames") as pbar:
        for idx, path in enumerate(paths):
            if not os.path.isfile(path):
                continue

            img = cv2.imread(path, cv2.IMREAD_COLOR)
            if img is None:
                print(f"Could not read image: {{path}}")
                continue

            try:
                # Process faces
                face_helper.clean_all()
                face_helper.read_image(img)
                face_helper.get_face_landmarks_5(only_center_face=False, resize=640, eye_dist_threshold=5)
                face_helper.align_warp_face()

                # Process each face
                restored_faces = []
                for i, cropped_face in enumerate(face_helper.cropped_faces):
                    try:
                        cropped_face_t = img2tensor(cropped_face / 255., bgr2rgb=True, float32=True)
                        normalize(cropped_face_t, (0.5, 0.5, 0.5), (0.5, 0.5, 0.5), inplace=True)
                        cropped_face_t = cropped_face_t.unsqueeze(0).to(device)

                        with torch.no_grad():
                            with torch.cuda.amp.autocast():
                                output = net(cropped_face_t, w=args.fidelity_weight, adain=True)[0]
                            restored_face = tensor2img(output, rgb2bgr=True, min_max=(-1, 1))
                            restored_faces.append(restored_face)
                    except Exception as e:
                        print(f"Error processing face: {{str(e)}}")
                        restored_faces.append(cropped_face)

                # Add all restored faces
                for restored_face in restored_faces:
                    face_helper.add_restored_face(restored_face)

                # Get final result
                if len(face_helper.restored_faces) > 0:
                    face_helper.get_inverse_affine(None)
                    restored_img = face_helper.paste_faces_to_input_image()
                else:
                    restored_img = img

                # Save the result
                save_path = os.path.join(args.output, 'final_results', os.path.basename(path))
                cv2.imwrite(save_path, restored_img)

                pbar.update(1)
                if (idx + 1) % 10 == 0:
                    print(f"Processed {{idx + 1}}/{{len(paths)}} images")

            except Exception as e:
                print(f"Error processing {{path}}: {{str(e)}}")
                cv2.imwrite(os.path.join(args.output, 'final_results', os.path.basename(path)), img)
                pbar.update(1)

if __name__ == "__main__":
    main()
'''
            f.write(script_content.format(weights_path, os.path.join(CODEFORMER_PATH, 'weights/retinaface_resnet50')))

        # Process frames with CodeFormer
        print("\n===== Processing frames with CodeFormer =====")
        process_cmd = f'python {inference_script} -i inputs/whole_imgs -o results -w {w} -s {upscale}'

        try:
            process = subprocess.Popen(
                process_cmd.split(),
                stdout=subprocess.PIPE,
                stderr=subprocess.STDOUT,
                universal_newlines=True,
                bufsize=1
            )

            with tqdm(total=saved_count, desc="Processing frames") as pbar:
                processed_count = 0
                while True:
                    output = process.stdout.readline()
                    if output == '' and process.poll() is not None:
                        break
                    if output:
                        if "Processed" in output:
                            try:
                                processed_count = int(output.split('/')[0].split()[-1])
                                pbar.n = processed_count
                                pbar.refresh()
                            except:
                                pass
                        print(output.strip())

            if process.returncode != 0:
                raise Exception(f"CodeFormer process failed with return code {process.returncode}")

            # Verify processed frames
            processed_frames = sorted(glob.glob(os.path.join('results/final_results', '*.png')))
            if not processed_frames:
                raise Exception("No processed frames found")

            print(f"\nFound {len(processed_frames)} processed frames")

            # Create video
            print("\nCreating output video...")
            temp_output = os.path.join(CODEFORMER_PATH, "temp_output.mp4")

            # Use ffmpeg for video creation
            output_fps = min(original_fps, target_fps)
            ffmpeg_cmd = (
                f'ffmpeg -framerate {output_fps} -i results/final_results/%08d.png '
                f'-c:v libx264 -preset medium -crf 23 -pix_fmt yuv420p {temp_output} -y'
            )
            # Run ffmpeg with error checking
            result = subprocess.run(ffmpeg_cmd, shell=True, capture_output=True, text=True)
            if result.returncode != 0:
                raise Exception(f"FFmpeg failed: {result.stderr}")

            if not os.path.exists(temp_output):
                raise Exception(f"FFmpeg did not create output file at {temp_output}")

            # Add audio if available
            if audio_clip is not None and os.path.exists(temp_audio_path):
                print("Adding audio...")
                final_output = os.path.join(CODEFORMER_PATH, "final_output.mp4")  # Use absolute path
                final_cmd = (
                    f'ffmpeg -i "{temp_output}" -i "{temp_audio_path}" '
                    f'-c:v copy -c:a aac -shortest "{final_output}" -y'
                )

                result = subprocess.run(final_cmd, shell=True, capture_output=True, text=True)
                if result.returncode != 0:
                    raise Exception(f"FFmpeg audio merge failed: {result.stderr}")

                # Verify final video
                final_video = VideoFileClip(final_output)
                final_duration = final_video.duration
                final_video.close()

                print(f"\nOriginal duration: {duration:.2f}s")
                print(f"Final duration: {final_duration:.2f}s")

                if os.path.exists(final_output):
                    shutil.copy2(final_output, output_path)
                    os.remove(final_output)
                else:
                    shutil.copy2(temp_output, output_path)
            else:
                shutil.copy2(temp_output, output_path)

            # Clean up temporary files
            for temp_file in [temp_output, temp_audio_path]:
                if os.path.exists(temp_file):
                    os.remove(temp_file)

            # Print statistics
            end_time = time.time()
            process_duration = end_time - start_time
            minutes = int(process_duration // 60)
            seconds = int(process_duration % 60)
            print(f"\nProcessing completed in {minutes} minutes and {seconds} seconds")
            print(f"Average processing speed: {saved_count/process_duration:.2f} frames/second")

            return output_path

        except Exception as e:
            print(f"Error during processing: {str(e)}")
            import traceback
            traceback.print_exc()
            return None

    finally:
        # Clean up and return to original directory
        try:
            # Clean up temporary directories
            for temp_dir in ['inputs/whole_imgs', 'results/final_results']:
                if os.path.exists(temp_dir):
                    shutil.rmtree(temp_dir)
        except Exception as e:
            print(f"Error cleaning up temporary files: {str(e)}")

        os.chdir(original_dir)
        clean_memory()
def clean_memory():
    """Clean up GPU and CPU memory"""
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

def run_wav2lip(input_video, input_audio, output_path):
    """Run Wav2Lip with correct parameters"""
    try:
        # Clean memory before processing
        clean_memory()

        # Change directory to WAV2LIP_PATH first
        original_dir = os.getcwd()
        os.chdir(WAV2LIP_PATH)

        # Construct the command with correct parameters
        command = [
            "python", "inference.py",
            "--checkpoint_path", "/content/Wav2Lip-GFPGAN/{wav2lipPath}/checkpoints/wav2lip.pth",
            "--face", input_video,
            "--audio", input_audio,
            "--outfile", output_path,
            "--wav2lip_batch_size", "4",
            "--face_det_batch_size", "4",
            "--resize_factor", "1"
        ]

        # Run the command with timeout
        process = subprocess.Popen(
            command,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )

        try:
            # Wait for completion with timeout
            stdout, stderr = process.communicate(timeout=300)  # 5 minute timeout

            if stdout:
                logger.info(f"Wav2Lip output: {stdout}")
            if stderr:
                logger.error(f"Wav2Lip error: {stderr}")

            success = process.returncode == 0

            # Check if output file exists
            if success and not os.path.exists(output_path):
                logger.error("Output file not created despite successful return code")
                success = False

            return success

        except subprocess.TimeoutExpired:
            process.kill()
            logger.error("Wav2Lip process timed out")
            return False

    except Exception as e:
        logger.error(f"Error running Wav2Lip: {str(e)}")
        return False
    finally:
        # Return to original directory
        os.chdir(original_dir)
        clean_memory()

def extract_faces(video_path, max_frames=35):
    """Extract face images from video."""
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    face_urls = []

    while cap.isOpened() and frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break

        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_detection.process(rgb_frame)

        if results.detections:
            for detection in results.detections:
                bboxC = detection.location_data.relative_bounding_box
                h, w, c = frame.shape
                x, y, w, h = int(bboxC.xmin * w), int(bboxC.ymin * h), int(bboxC.width * w), int(bboxC.height * h)

                if x >= 0 and y >= 0 and w > 0 and h > 0:
                    face = frame[y:y+h, x:x+w]
                    face_filename = os.path.join(FACE_FRAMES_FOLDER, f"face_{frame_count}.jpg")
                    cv2.imwrite(face_filename, face)
                    face_urls.append(f"/frames/face_{frame_count}.jpg")
                    frame_count += 1

        if frame_count >= max_frames:
            break

    cap.release()
    return face_urls

class LipSyncGenerator:
    """Class to process lip landmarks"""
    def __init__(self):
        self.face_mesh = mp.solutions.face_mesh.FaceMesh(
            static_image_mode=False, max_num_faces=1, min_detection_confidence=0.5, min_tracking_confidence=0.5
        )
        self.lip_indices = [
            61, 146, 91, 181, 84, 17, 314, 405, 321, 375,
            291, 308, 324, 318, 402, 317, 14, 87, 178, 88,
            95, 78, 191, 80, 81, 82
        ]

    def extract_face_landmarks(self, frame):
        """Extract facial landmarks using MediaPipe"""
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = self.face_mesh.process(rgb_frame)

        if results.multi_face_landmarks:
            face_landmarks = results.multi_face_landmarks[0]
            frame_height, frame_width = frame.shape[:2]
            lip_points = np.array([
                (int(face_landmarks.landmark[idx].x * frame_width), int(face_landmarks.landmark[idx].y * frame_height))
                for idx in self.lip_indices
            ], dtype=np.float32)
            return lip_points
        return None

    def process_lip_tracking(self, video_path, output_path):
        """Process video and map lip movements"""
        print(f"Processing video: {video_path}")
        video_clip = VideoFileClip(video_path)
        frames = []

        for i in tqdm(range(int(video_clip.duration * video_clip.fps)), desc="Processing Frames"):
            frame = video_clip.get_frame(i / video_clip.fps)
            if frame.dtype != np.uint8:
                frame = (frame * 255).astype(np.uint8)

            # Create a copy of the frame before drawing
            frame_copy = frame.copy()

            landmarks = self.extract_face_landmarks(frame)
            if landmarks is not None:
                for point in landmarks:
                    x, y = map(int, point)
                    cv2.circle(frame_copy, (x, y), 3, (0, 255, 0), -1)

            frames.append(frame_copy)

        output_clip = ImageSequenceClip(frames, fps=video_clip.fps)
        output_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
        print(f"✅ Lip-mapped video saved: {output_path}")
        return True

lip_sync = LipSyncGenerator()


@app.route("/process_video", methods=["POST"])
def process_video():
    try:
        global global_timestamp
        if global_timestamp is None:
            return jsonify({"error": "No video has been uploaded yet!"}), 400

        timestamp = int(time.time())
        input_video_path = os.path.join(INPUT_FACE_VIDEO, f"input_video_{global_timestamp}.mp4")
        lip_video_path = os.path.join(PROCESSED_VIDEOS_FOLDER, f"lip_mapped_{timestamp}.mp4")

        # Check if input video exists
        if not os.path.exists(input_video_path):
            return jsonify({"error": "Input video not found. Please process a video first."}), 400


        print("Checking video file before processing...")
        print("File exists:", os.path.exists(input_video_path))
        print("File size:", os.path.getsize(input_video_path) if os.path.exists(input_video_path) else "N/A")

        # Extract face frames and get base64 encoded images
        face_frames_data = []
        face_urls = extract_faces(input_video_path)

        for i in range(len(face_urls)):
            frame_path = os.path.join(FACE_FRAMES_FOLDER, f"face_{i}.jpg")
            if os.path.exists(frame_path):
                with open(frame_path, "rb") as img_file:
                    base64_image = base64.b64encode(img_file.read()).decode('utf-8')
                    face_frames_data.append(f"data:image/jpeg;base64,{base64_image}")

        # Process video for lip tracking
        success = lip_sync.process_lip_tracking(input_video_path, lip_video_path)
        if not success:
            return jsonify({"error": "Lip mapping failed"}), 500


        # Convert video to base64
        with open(lip_video_path, "rb") as video_file:
            base64_video = base64.b64encode(video_file.read()).decode('utf-8')

        response = {
            "face_frames": face_frames_data,
            "lip_mapped_video": f"data:video/mp4;base64,{base64_video}"
        }

        return jsonify(response)

    except Exception as e:
        print(f"Error in process_video: {str(e)}")
        return jsonify({"error": str(e)}), 500
    finally:
        clean_memory()

        # Only delete video after processing
        if os.path.exists(input_video_path):
            try:
                os.remove(input_video_path)
            except Exception as cleanup_error:
                print(f"Error cleaning up input video file: {cleanup_error}")

        if os.path.exists(lip_video_path):
            try:
                os.remove(lip_video_path)
            except Exception as cleanup_error:
                print(f"Error cleaning up processed video file: {cleanup_error}")

@app.route("/face_frames/<filename>")
def get_face_frame(filename):
    try:
        file_path = os.path.join(FACE_FRAMES_FOLDER, filename)
        if not os.path.exists(file_path):
            return jsonify({"error": "File not found"}), 404

        # Read the image and convert to base64
        with open(file_path, "rb") as image_file:
            encoded_string = base64.b64encode(image_file.read()).decode('utf-8')

        return jsonify({
            "image": encoded_string,
            "content_type": "image/jpeg"
        })
    except Exception as e:
        print(f"Error serving face frame {filename}: {str(e)}")
        return jsonify({"error": str(e)}), 500

    finally:
        # Delete the file after sending
        if os.path.exists(file_path):
            os.remove(file_path)

@app.route("/processed_videos/<filename>")
def get_processed_video(filename):
    try:
        file_path = os.path.join(PROCESSED_VIDEOS_FOLDER, filename)
        if not os.path.exists(file_path):
            return jsonify({"error": "File not found"}), 404

        # Read the video and convert to base64
        with open(file_path, "rb") as video_file:
            encoded_string = base64.b64encode(video_file.read()).decode('utf-8')

        return jsonify({
            "video": encoded_string,
            "content_type": "video/mp4"
        })
    except Exception as e:
        print(f"Error serving video {filename}: {str(e)}")
        return jsonify({"error": str(e)}), 500

    finally:
        # Delete the file after sending
        if os.path.exists(file_path):
            os.remove(file_path)

@app.route('/health', methods=['GET'])
def health_check():
    """Simple health check endpoint"""
    return jsonify({"status": "healthy"})

@app.route('/download_youtube', methods=['POST'])
def download_youtube():
    """Download and trim a YouTube video"""
    try:
        data = request.json
        youtube_url = data.get('youtube_url')
        start_time = data.get('start_time')
        duration = data.get('duration')
        print(f"Received URL: {youtube_url}")  # Debug
        print("Received data:", request.json)

        if not youtube_url or start_time is None or duration is None:
            return jsonify({"error": "Missing required parameters"}), 400

        timestamp = int(time.time())
        raw_video_path = os.path.join(OUTPUT_PATH, f"raw_video_{timestamp}.mp4")
        trimmed_video_path = os.path.join(OUTPUT_PATH, f"trimmed_video_{timestamp}.mp4")

        yt_command = [
            "yt-dlp",
            "-f", "bestvideo+bestaudio/best",  # This ensures high-quality video & audio
            "--merge-output-format", "mp4",  # Force MP4 format
            "-o", raw_video_path,
            youtube_url
        ]


        yt_process = subprocess.run(yt_command, capture_output=True, text=True)
        if yt_process.returncode != 0:
            return jsonify({"error": "Failed to download YouTube video", "details": yt_process.stderr}), 500

        # Verify file existence
        if not os.path.exists(raw_video_path) or os.stat(raw_video_path).st_size == 0:
            return jsonify({"error": "Downloaded video file is missing or empty"}), 500

        # Trim video using FFmpeg with debugging
        trim_command = [
            "ffmpeg",
            "-i", raw_video_path,  # Load input first
            "-ss", str(start_time),  # Seek after input to avoid black frames
            "-t", str(duration),
            "-c:v", "libx264", "-preset", "fast", "-crf", "23",  # Re-encode video
            "-c:a", "aac", "-b:a", "128k",  # Ensure audio is encoded
            "-movflags", "+faststart",  # Optimize playback
            trimmed_video_path
        ]

        trim_process = subprocess.run(trim_command, capture_output=True, text=True)
        if trim_process.returncode != 0:
            return jsonify({"error": "Failed to trim video", "details": trim_process.stderr}), 500

        return send_file(trimmed_video_path, mimetype='video/mp4', as_attachment=True, download_name="trimmed_video.mp4")

    except Exception as e:
        return jsonify({"error": str(e)}), 500

    finally:
        # Safely delete raw and trimmed videos
        for file_path in [raw_video_path, trimmed_video_path]:
            if file_path and os.path.exists(file_path):
                os.remove(file_path)


@app.route("/convert_fps", methods=["POST"])
def convert_fps():
    """Convert video to the specified frame rate."""
    try:
        if "video" not in request.files:
            return jsonify({"error": "No video file uploaded"}), 400

        video = request.files["video"]
        frame_rate = request.form.get("frame_rate")

        if not frame_rate:
            return jsonify({"error": "Frame rate not specified"}), 400

        timestamp = int(time.time())
        input_video_path = os.path.join(OUTPUT_PATH, f"input_video_{timestamp}.mp4")
        output_video_path = os.path.join(OUTPUT_PATH, f"converted_video_{timestamp}.mp4")

        # Save input video
        video.save(input_video_path)

        # Run FFmpeg to change FPS
        ffmpeg_command = [
            "ffmpeg", "-i", input_video_path,
            "-filter:v", f"fps={frame_rate}",
            "-c:a", "copy",  # Copy audio without re-encoding
            output_video_path
        ]

        process = subprocess.run(ffmpeg_command, capture_output=True, text=True)

        if process.returncode != 0:
            return jsonify({"error": "FFmpeg failed", "details": process.stderr}), 500

        return send_file(output_video_path, mimetype="video/mp4", as_attachment=True, download_name="converted_video.mp4")

    except Exception as e:
        return jsonify({"error": str(e)}), 500
    finally:
        # Delete both videos safely
        for file_path in [input_video_path, output_video_path]:
            if file_path and os.path.exists(file_path):
                os.remove(file_path)
        clean_memory()

@app.route("/text_to_speech", methods=["POST"])
def text_to_speech():
    """Convert text to speech and return audio file for download."""
    try:
        data = request.json
        text = data.get("text")
        language = data.get("language", "en")  # Default to English

        if not text:
            return jsonify({"error": "No text provided"}), 400

        # Convert text to speech using gTTS
        tts = gTTS(text=text, lang=language)
        audio_buffer = io.BytesIO()
        tts.write_to_fp(audio_buffer)
        audio_buffer.seek(0)

        return send_file(audio_buffer, mimetype="audio/mp3", as_attachment=True, download_name="speech.mp3")

    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route('/process', methods=['POST'])
def process():
    """Process video and audio files with Wav2Lip"""
    try:

        global global_timestamp
        # Check if files are provided
        if 'video' not in request.files or 'audio' not in request.files:
            return jsonify({"error": "Missing video or audio file"}), 400

        video = request.files['video']
        audio = request.files['audio']

        # Validate file types
        if not video.filename.lower().endswith(('.mp4', '.avi', '.mov')):
            return jsonify({"error": "Invalid video format. Please use MP4, AVI, or MOV"}), 400

        if not audio.filename.lower().endswith(('.wav', '.mp3')):
            return jsonify({"error": "Invalid audio format. Please use WAV or MP3"}), 400

        # Create unique filenames
        timestamp = int(time.time())
        global_timestamp = timestamp
        input_video_path = os.path.join(OUTPUT_PATH, f'input_video_{timestamp}.mp4')
        input_audio_path = os.path.join(OUTPUT_PATH, f'input_audio_{timestamp}.wav')
        output_path = os.path.join(OUTPUT_PATH, f'result_{timestamp}.mp4')
        face_detect_input_path = os.path.join(INPUT_FACE_VIDEO, f'input_video_{timestamp}.mp4')

        # Save files before
        # video.save(input_video_path)
        # audio.save(input_audio_path)

        # Save files
        video.save(input_video_path)
        video.stream.seek(0)  # Reset stream position
        with open(face_detect_input_path, "wb") as f:
            f.write(video.read())  # Write file explicitly

        time.sleep(2)  # Give system some time to complete file write
        audio.save(input_audio_path)

        try:
            # Check file sizes
            video_size = os.path.getsize(input_video_path)
            audio_size = os.path.getsize(input_audio_path)

            # Log file info
            logger.info(f"Video size: {video_size / (1024*1024):.2f} MB")
            logger.info(f"Audio size: {audio_size / (1024*1024):.2f} MB")

            # Run processing
            success = run_wav2lip(input_video_path, input_audio_path, output_path)

            if not success:
                return jsonify({
                    "error": "Failed to process video. Please check the input files and try again."
                }), 500

            if not os.path.exists(output_path):
                return jsonify({"error": "Output file not generated"}), 500

            return send_file(
                output_path,
                mimetype='video/mp4',
                as_attachment=True,
                download_name='result.mp4'
            )

        finally:
            # Clean up input files
            try:
                os.remove(input_video_path)
                os.remove(input_audio_path)
            except Exception as e:
                logger.warning(f"Error cleaning up input files: {e}")

    except Exception as e:
        logger.error(f"Error in process endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500
    finally:
        clean_memory()

@app.route('/enhance', methods=['POST'])
def enhance_video():
    """Enhance the latest Wav2Lip output video using CodeFormer"""
    try:
        global global_timestamp

        if global_timestamp is None:
            return jsonify({"error": "No video has been processed yet"}), 404

        # Get the processed video path using the stored timestamp
        input_video = os.path.join(OUTPUT_PATH, f'result_{global_timestamp}.mp4')

        if not os.path.exists(input_video):
            return jsonify({"error": "Processed video not found"}), 404

        # Set output path for enhanced video
        timestamp = int(time.time())
        enhanced_output = os.path.join(OUTPUT_PATH, f'enhanced_result_{timestamp}.mp4')

        logger.info(f"Enhancing video: {input_video}")
        logger.info(f"Output path: {enhanced_output}")

        # Process video with CodeFormer
        final_path = process_enhanced_video(input_video, enhanced_output)

        if not final_path or not os.path.exists(final_path):
            logger.error("Enhancement failed - no output file generated")
            return jsonify({"error": "Failed to enhance video"}), 500

        # Return enhanced video
        return send_file(
            final_path,
            mimetype='video/mp4',
            as_attachment=True,
            download_name='enhanced_result.mp4'
        )

    except Exception as e:
        logger.error(f"Error in enhance endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500
    finally:
        clean_memory()

def start_ngrok():
    """Start ngrok using subprocess"""
    try:
        # Kill any existing ngrok processes
        subprocess.run('pkill ngrok', shell=True, capture_output=True)
        time.sleep(2)

        # Start ngrok in background
        subprocess.Popen(['ngrok', 'http', '5000'])
        time.sleep(3)  # Wait for ngrok to start

        # Get the public URL
        import requests
        response = requests.get('http://localhost:4040/api/tunnels')
        data = response.json()
        url = data['tunnels'][0]['public_url']
        print(f"\n{'='*50}")
        print(f"Ngrok URL: {url}")
        print("Copy this URL into the 'Server URL' field in the HTML page")
        print(f"{'='*50}\n")
        return url

    except Exception as e:
        print(f"Error starting ngrok: {e}")
        return None

if __name__ == "__main__":
    # Start ngrok first
    ngrok_url = start_ngrok()
    if not ngrok_url:
        print("Warning: Failed to get ngrok URL, but continuing with local server")

    # Run Flask app
    app.run(host='0.0.0.0', port=5000)
```

# NGROK

* Issue with ngrok
Your start_ngrok function assumes ngrok is installed and running. However:

* Killing existing ngrok processes with pkill ngrok may not work if ngrok is not running.
* If requests.get('http://localhost:4040/api/tunnels') fails, the script will break.
✅ Fix: Instead of subprocess.run('pkill ngrok', shell=True), use:

`subprocess.run(["pkill", "-f", "ngrok"], shell=True, capture_output=True)`
